```python
import numpy as np
import matplotlib.pyplot as plt

# N√∫mero de amostras
n_amostras = 10000
amostras = []

# Gerar amostras usando o m√©todo da aceita√ß√£o-rejei√ß√£o
while len(amostras) < n_amostras:
    U1 = np.random.uniform(0, 1)  # Gerar U1
    U2 = np.random.uniform(0, 1)  # Gerar U2

    if U2 <= U1**2:  # Crit√©rio de aceita√ß√£o
        amostras.append(U1)

amostras = np.array(amostras)

# Gerar valores para a PDF te√≥rica
x_vals = np.linspace(0, 1, 1000)
pdf_vals = 3 * x_vals**2

# Plotar histograma e PDF te√≥rica
plt.figure(figsize=(8, 5))
plt.hist(amostras, bins=30, density=True, alpha=0.6, color='b', label="Histograma normalizado")
plt.plot(x_vals, pdf_vals, 'r-', lw=2, label="PDF anal√≠tica $f(x) = 3x^2$")
plt.xlabel("x")
plt.ylabel("Densidade de probabilidade")
plt.title("Histograma de amostras vs. PDF anal√≠tica")
plt.legend()
plt.grid()
plt.show()
```
